### Sample metadata and TPM

```r
### Gene Info file from Gencode v34lift37
geneinfo = fread("../tables/gene_info.txt", sep = "\t", header = TRUE, data.table = FALSE)

### Supplemental Table 1b
rna_meta = fread("../tables/TableS1b.txt",sep="\t",data.table=F)

### TPM on Figshare - filter for 8 samples in this study
tpm_path="/projects/CARDIPS/analysis/ifng/pipeline/rna/tpm.txt"
tpm = fread(tpm_path,sep="\t",data.table=F)
rownames(tpm) = tpm$V1 
tpm$V1 = NULL
```

```r
# get expressed genes
expressed_tpm = function(tpm, min_tpm, min_samples = 0.2)
{
    expressed = as.matrix(tpm)

    expressed[as.matrix(tpm) <  min_tpm] = 0
    expressed[as.matrix(tpm) >= min_tpm] = 1

    tpm_f = tpm[rowSums(expressed) >= (min_samples * ncol(tpm)),]
    
    return(tpm_f)
}

ifng_tpm = expressed_tpm(tpm, 1, 0.2)


ifng_tpm$gene_name = suppressMessages(mapvalues(rownames(ifng_tpm), geneinfo$gene_id, geneinfo$gene_name))

genes_unique = data.frame(table(ifng_tpm$gene_name)) %>% filter(Freq == 1) %>% mutate(Var1 = as.vector(Var1)) %>% pull(Var1)

ifng_tpm = ifng_tpm %>% filter(gene_name %in% genes_unique) 
rownames(ifng_tpm) = ifng_tpm$gene_name
ifng_tpm$gene_name = NULL
```

### Process Tabula Sapiens scRNA and Find Cell type Markers

```r
scdata = ts_heart
expr = ifng_tpm

Idents(scdata)        = "cell_ontology_class"
scdata <- subset(scdata, subset = cell_ontology_class != "macrophage")
scdata <- subset(scdata, subset = cell_ontology_class != "hepatocyte")

scdata <- NormalizeData(scdata)
scdata <- FindVariableFeatures(scdata, selection.method = "vst", nfeatures = 2000)
all.genes <- rownames(scdata)
scdata <- ScaleData(scdata, features = all.genes)
scdata <- RunPCA(scdata, features = VariableFeatures(object = scdata))


scdata <- FindNeighbors(scdata, dims = 1:10)
scdata <- FindClusters(scdata, resolution = 0.08)

scdata <- RunUMAP(scdata, dims = 1:10)
DimPlot(scdata, reduction = "umap")

scdata@meta.data$RNA_snn_res.0.08 = suppressMessages(mapvalues(scdata@meta.data$RNA_snn_res.0.08, from = c(0:5), to = c("cardiac_muscle",
                                                                                                                "cardiac_muscle", "endothelial",
                                                                                                                    "fibroblast",
                                                                                                                "smooth_muscle", "unknown")))

table(scdata$RNA_snn_res.0.08)

Idents(scdata) = "RNA_snn_res.0.08"
DimPlot(scdata, reduction = "umap")

scdata <- subset(scdata, subset = RNA_snn_res.0.08 != "unknown")

markers               = FindAllMarkers(scdata, only.pos = TRUE, min.pct = 0.2, logfc.threshold = 0.1)

```

### Make signature matrix 

```r
markers               = markers[ markers$p_val_adj < 0.05 & markers$gene %in% rownames(ifng_tpm),]

markers = markers %>% group_by(cluster) %>% top_n(250, avg_log2FC)

markers = markers[ markers$gene %in% names(table(markers$gene))[ table(markers$gene)  == 1],]
markers = markers %>% group_by(cluster) %>% top_n(100, avg_log2FC)
table(markers$cluster)
genes2cibersort       = unique(markers$gene)
sig.avg_exp           = AverageExpression(scdata, assays = "RNA", features = unique(genes2cibersort))
sig.avg_exp           = sig.avg_exp$RNA
colnames(sig.avg_exp) = gsub(" ", "_", colnames(sig.avg_exp))
```

### Deconvolute

```r
source("cibersort.R" )
cibersort             = CIBERSORT(sig.avg_exp, ifng_tpm, QN = FALSE, perm = 0)
cibersort             = as.data.frame(cibersort)
cibersort$id = rownames(cibersort)
cibersort
```
