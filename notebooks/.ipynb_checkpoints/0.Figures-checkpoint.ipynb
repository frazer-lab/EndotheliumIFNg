### Artery and CM Centroids

```r
pcadata2$TissueType = ifelse(pcadata2$Tissue %in% c("Coronary","Aorta"), "Artery",
                            ifelse(pcadata2$Tissue %in% c("Ventricle","Atrium"), "CM","CVPC"))
centroid_cardiomyocytes = colMeans(pcadata2[ pcadata2$TissueType == "CM", 2:3])
centroid_arteries = colMeans(pcadata2[ pcadata2$TissueType == "Artery", 2:3])

ifng_pcs = pcadata2[ pcadata2$treatment %in% c("IFNg","CONTROL"),]

centroid_cardiomyocytes = colMeans(pcadata2[ pcadata2$TissueType == "CM", 2:3])
centroid_arteries = colMeans(pcadata2[ pcadata2$TissueType == "Artery", 2:3])
centroid_cardiomyocytes

euclidean_distance = function(PC1, PC2, centroid) {
  sqrt(sum((c(PC1, PC2) - centroid)^2))
}

ifng_pcs$distance_to_cardiomyocytes = mapply(euclidean_distance,ifng_pcs$PC1,ifng_pcs$PC2,
                                             MoreArgs = list(centroid = centroid_cardiomyocytes))
ifng_pcs$distance_to_arteries = mapply(euclidean_distance, ifng_pcs$PC1,ifng_pcs$PC2,
                                             MoreArgs = list(centroid = centroid_arteries))

ifng_pcs_wide = reshape(ifng_pcs[, c("udid","treatment","distance_to_cardiomyocytes","distance_to_arteries"),], idvar = "udid", timevar = "treatment", direction = "wide")


t.test(ifng_pcs_wide$distance_to_arteries.IFNg - ifng_pcs_wide$distance_to_cardiomyocytes.IFNg,
       ifng_pcs_wide$distance_to_arteries.CONTROL - ifng_pcs_wide$distance_to_cardiomyocytes.CONTROL,
       paired = TRUE)
```

### Chromatin State Annotations

```r
### Collapse States

### path to Epimap chromatin state metadata
chrom_states           = fread(epimap_state_metadata_file, sep = "\t", header = TRUE, data.table = FALSE)
colnames(chrom_states) = c("state", "mnemonic", "description", "color", "code")
rownames(chrom_states) = chrom_states$mnemonic

chrom_states$collapsed [ grepl("Tss",chrom_states$mnemonic)] = "Promoter"
chrom_states$collapsed [ grepl("Tx",chrom_states$mnemonic)] = "Transcribed"
chrom_states$collapsed [ grepl("Enh",chrom_states$mnemonic)] = "Enhancer"
chrom_states$collapsed [ chrom_states$mnemonic %in% c("ReprPC","ReprPCWk","Het","ZNF/Rpts")] = "Repressed"
chrom_states$collapsed [ chrom_states$mnemonic == "Quies"] = "Quiescent"
chrom_states$collapsed [ grepl("Biv",chrom_states$mnemonic)] = "Bivalent"

chrom_states$collapsed_color [ chrom_states$collapsed == "Promoter" ] = "red"
chrom_states$collapsed_color [ chrom_states$collapsed == "Transcribed" ] = "green"
chrom_states$collapsed_color [ chrom_states$collapsed == "Enhancer" ] = "yellow"
chrom_states$collapsed_color [ chrom_states$collapsed == "Repressed" ] = "grey"
chrom_states$collapsed_color [ chrom_states$collapsed == "Quiescent" ] = "white"
chrom_states$collapsed_color [ chrom_states$collapsed == "Bivalent" ] = "darkkhaki"
chrom_states$new_state = c(rep(1,4),rep(3,2),rep(2,5),4,4,5,5,4,4,6)

state_key = unique(chrom_states[,c("mnemonic","collapsed")])

epimap.dir = path_to_epimap_chromatin_states
epimap.out.dir = path_to_collapsed_epimap_out

epimap.fs = list.files(epimap.dir, full.names=T)


writeEpimapBeds = function(file) {
    fname = str_extract(file,"BSS[0-9]+")
    chrom_df = merge(fread(file, sep = "\t", header = FALSE, data.table = FALSE)[,1:4],
                             state_key, by.x="V4",by.y="mnemonic")[,2:5] 
    chrom_df$start = chrom_df$V2 + 1
    chrom_df$bss = fname
    fwrite(chrom_df %>% select(V1, start, V3, collapsed, bss) %>% arrange(V1,start,V3),
          paste0(epimap.out.dir,"/",fname,".bed"),
               sep="\t",row.names=F,col.names=F,quote=F)


}
lapply(epimap.fs,writeEpimapBeds)

### use bedtools intersect with peak summits and collapsed chromatin states to annotate peaks
pseudo_cmd = "bedtools -a summit.bed -b $collapsed_chromatin_file -wb > $out_file_name"
```

```r
ann_path=path_to_annotated_peak_fs
annotated.fs = list.files(ann_path, full.names=T)

### Test Chromatin State enrichment with Fisher's Exact test
run_fisher = as.data.frame(rbindlist(lapply(annotated.fs, function(x){
    df = fread(x, sep="\t",data.table=F)
    df$upreg = ifelse(df$V9 %in% upreg$peak_id, 1, 0)
    by_state = rbindlist(lapply(unique(df$V4), function(state){
        tab = as.data.frame(table(df$V4,df$upreg))
        tab$Var1 = as.character(tab$Var1)
        tab$Var2 = as.character(tab$Var2)
        
        statedf = tab[ tab$Var1 == state ,]
        x11 = statedf$Freq [ statedf$Var2 == 1]
        x12 = statedf$Freq [ statedf$Var2 == 0]
        notstate = tab[ tab$Var1 != state ,]
        x21 = sum(notstate$Freq [ notstate$Var2 == 1])
        x22 = sum(notstate$Freq [ notstate$Var2 == 0])
        fishtest = fisher.test(matrix(c(x11, x12, x21, x22), nrow =2 ))
        out = data.frame(State = state, x11 = x11, x12 = x12, x21 = x21, x22 = x22,
                         OR = fishtest$estimate, Pvalue = fishtest$p.value)
        return(out)
    }))
    by_state$bss = str_extract(x, "BSS[0-9]+")
    return(by_state)
})))
```
